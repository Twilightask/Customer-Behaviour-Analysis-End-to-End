```python
# Since there are nulls in review ratings so we should address them using median instead of the mean because the median is less affected by outliers
# compared to mean,
# also it is advised to take median for every category separately instead of taking an overall median.
```

```python
df[['Category','Review Rating']]
```

```python
def fillna_median(x):
    return x.fillna(x.median())
```

```python
df["Review Rating"] = df.groupby("Category")['Review Rating'].transform(fillna_median)
```

```python
# transform() = do something per group, but give a value for every row.
```

```python
df.isnull().sum()
```

```python
# There could be one problem while doing analysis in column names as there contains spaces in between columns and column names are inconsistent.
```

```python
# So we will convert it to snake casing
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```

```python
# .str is used because we are applying a string operation to many column names at once, not to a single string.
```

```python
# Feature engineering
```

```python
# Adding age groups column
labels = ['Young_adult','adult','middle_aged','senior']
df['age_group'] = pd.qcut(df['age'], q = 4, labels = labels) 
```

```python
df[['age_group','age']].reset_index(drop = True)
```

```python
df['frequency_of_purchases'].unique()
```

```python
frequency_mapping = {
    "Fortnightly":14,
    'Weekly':7,
    'Annually':365,
    'Quarterly':90,
    'Bi-Weekly':14
}

df['purchase_frequency_days'] = df["frequency_of_purchases"].map(frequency_mapping)
```
